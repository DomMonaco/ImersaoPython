```python
import pandas as bibliotecaPandas

import plotly.express as bibliotecaPlotlyExpress
```

```python
arquivoExcelPrincipal = bibliotecaPandas.read_excel("/content/[Faça uma cópia para editar] Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")

arquivoExcelPrincipal.head(10)
```

```python
arquivoExcelTotalAcoes = bibliotecaPandas.read_excel("/content/[Faça uma cópia para editar] Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")

arquivoExcelTotalAcoes.head(10)
```

```python
arquivoExcelTicker = bibliotecaPandas.read_excel("/content/[Faça uma cópia para editar] Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")

arquivoExcelTicker.head(10)
```

```python
arquivoExcelPrincipal = arquivoExcelPrincipal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal = arquivoExcelPrincipal.rename(columns={'Último (R$)':'valorFinal', 'Var. Dia (%)':'valorDiaPercentual'}).copy()

arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal['valorPorcentual'] = arquivoExcelPrincipal['valorDiaPercentual'] / 100
arquivoExcelPrincipal['valorInicial'] = arquivoExcelPrincipal['valorFinal'] / (arquivoExcelPrincipal['valorPorcentual'] + 1)

arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal = arquivoExcelPrincipal.merge(arquivoExcelTotalAcoes, left_on='Ativo', right_on='Código', how='left')
arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal = arquivoExcelPrincipal.drop(columns=['Código'])
arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal['variacaoReais'] = (arquivoExcelPrincipal['valorFinal'] - arquivoExcelPrincipal['valorInicial']) * arquivoExcelPrincipal['Qtde. Teórica']

arquivoExcelPrincipal
```

```python
bibliotecaPandas.options.display.float_format = '{:.2f}'.format
```

```python
arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal['Qtde. Teórica'] = arquivoExcelPrincipal['Qtde. Teórica'].astype(int)
arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal = arquivoExcelPrincipal.rename(columns={'Qtde. Teórica':'quantidadeTeorica'}).copy()

arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal['resultado'] = arquivoExcelPrincipal['variacaoReais'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))

arquivoExcelPrincipal
```

```python
arquivoExcelPrincipal = arquivoExcelPrincipal.merge(arquivoExcelTicker, left_on='Ativo', right_on='Ticker', how='left')
arquivoExcelPrincipal = arquivoExcelPrincipal.drop(columns=['Ticker'])
arquivoExcelPrincipal
```

```python
# Calculando o maior valor
maior = arquivoExcelPrincipal['variacaoReais'].max()

# Calculando o menor valor
menor = arquivoExcelPrincipal['variacaoReais'].min()

# Calculando a média
media = arquivoExcelPrincipal['variacaoReais'].mean()

# Calculando a média de quem subiu
media_subiu = arquivoExcelPrincipal[arquivoExcelPrincipal['resultado'] == 'Subiu']['variacaoReais'].mean()

# Calculando a média de quem desceu
media_desceu = arquivoExcelPrincipal[arquivoExcelPrincipal['resultado'] == 'Desceu']['variacaoReais'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
arquivoExcelPrincipalSubiu = arquivoExcelPrincipal[arquivoExcelPrincipal['resultado'] == 'Subiu']
arquivoExcelPrincipalSubiu
```

```python
arquivoExcelAnaliseSegmento = arquivoExcelPrincipalSubiu.groupby('Nome')['variacaoReais'].sum().reset_index()

arquivoExcelAnaliseSegmento
```

```python
arquivoExcelAnaliseSaldo = arquivoExcelPrincipal.groupby('resultado')['variacaoReais'].sum().reset_index()
arquivoExcelAnaliseSaldo
```

```python
grafico = bibliotecaPlotlyExpress.bar(arquivoExcelAnaliseSaldo, x='resultado', y='variacaoReais', text='variacaoReais', title='Variação Reais por Resultado')
grafico.show()
```
